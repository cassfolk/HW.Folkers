```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import os
import glob

import requests
import numpy as np
```

## Get info from station_status.json

```python
# make URL & pull json
url_station_status = "https://gbfs.citibikenyc.com/gbfs/en/station_status.json"
response_station_status = requests.get(url_station_status).json()
```

```python
response_station_status
```

```python
data = response_station_status['data']['stations']
```

```python
station_status_df = pd.DataFrame(data)
station_status_df.head()
```

## Get info from station_information.json

```python
# make URL & pull json
url_station_information = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
response_station_information = requests.get(url_station_information).json()
```

```python
response_station_information
```

```python
data_info = response_station_information['data']['stations']
```

```python
station_info_df = pd.DataFrame(data_info)
station_info_df.head()
```

## CSV both DFs (to then be put into Tableau)

```python
file_outpath_1 = "station_status.csv"

station_status_df.to_csv(file_outpath_1)
station_status_df.head(2)
```

```python
file_outpath_2 = "station_info.csv"

station_info_df.to_csv(file_outpath_2)
station_info_df.head(2)
```

## Merge CSVs into year DF-->new CSV

##### 2019

```python
path = 'Month_csv/2019'
all_2019 = glob.glob(path + "/*.csv")

df_all_2019 = pd.concat((pd.read_csv(f) for f in all_2019))
df_all_2019.head()
```

```python
df_all_2019.describe()
```

```python
file_outpath_2019 = "all_2019.csv"

df_all_2019.to_csv(file_outpath_2019)
```
